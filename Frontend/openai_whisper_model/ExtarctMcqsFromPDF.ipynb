```python
!pip install pdfplumber
import pdfplumber
import pandas as pd

def extract_mcqs_from_pdf(pdf_path):
    mcqs = []

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            print(text)
            mcqs.append(text)

    return mcqs

import re
import pandas as pd

def save_mcqs_to_excel(mcqs, excel_path):
    # Format the MCQs
    formatted_mcqs = []
    for mcq in mcqs:
        # Remove extra spaces and newlines
        mcq = re.sub(r'\s+', ' ', mcq.strip())
        # Add line breaks before options (assuming options start with capital letters)
        mcq = re.sub(r'([A-Z])', r'\n\1', mcq)
        # print(mcq)
        formatted_mcqs.append(mcq)

    # Create a DataFrame from the formatted MCQs
    df = pd.DataFrame({'MCQs': formatted_mcqs})

    # Save the DataFrame to an Excel file
    df.to_excel(excel_path, index=False)





# Path to the PDF file
pdf_file = '/content/p.pdf'

# Path to the Excel file to save the MCQs
excel_file = '/content/B.xlsx'

# Extract MCQs from the PDF
extracted_mcqs = extract_mcqs_from_pdf(pdf_file)

# Save the MCQs to an Excel file
save_mcqs_to_excel(extracted_mcqs, excel_file)
# !pip install pdfplumber
# !pip install pdfplumber
# import pdfplumber
# import pandas as pd

# def extract_mcqs_from_pdf(pdf_path):
#     mcqs = []
#     start_extraction = False

#     with pdfplumber.open(pdf_path) as pdf:
#         for page in pdf.pages:
#             text = page.extract_text()

#             # Start extracting MCQs after the specified line
#             if "Enzynies that catalyze the transfer of groups within molecule are" in text:
#                 start_extraction = True

#             # Extract MCQs
#             if start_extraction:
#                 mcqs.append(text.strip())
#                 print(mcqs)

#     return mcqs

# def save_mcqs_to_excel(mcqs, excel_path):
#     # Create a DataFrame from the MCQs list
#     df = pd.DataFrame({'MCQs': mcqs})

#     # Save the DataFrame to an Excel file
#     df.to_excel(excel_path, index=False)

# # Path to the PDF file
# pdf_file = '/content/c.pdf'

# # Path to the Excel file to save the MCQs
# excel_file = '/content/B.xlsx'

# # Extract MCQs from the PDF
# extracted_mcqs = extract_mcqs_from_pdf(pdf_file)

# # Save the MCQs to an Excel file
# save_mcqs_to_excel(extracted_mcqs, excel_file)

```
