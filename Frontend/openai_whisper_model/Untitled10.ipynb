```python
# !pip install transformers
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np
import pandas as pd

from transformers import DistilBertTokenizer, TFDistilBertForSequenceClassification

from sklearn.cluster import KMeans
import numpy as np

input_file = '/content/my3.xlsx'
df = pd.read_excel(input_file)
column_name = 'New Column'

# Retrieve the data from the desired column
mcqs = df.iloc[:, 0].tolist()

```

```python
#  this is the orignal code for distilbert
from transformers import DistilBertTokenizer, DistilBertModel
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
from openpyxl import load_workbook
# Load the existing Excel file

# Initialize the DistilBERT tokenizer and model
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
model = DistilBertModel.from_pretrained('distilbert-base-uncased')



# Encode the MCQs using DistilBERT
encoded_inputs = tokenizer(mcqs, padding=True, truncation=True, return_tensors='pt')
outputs = model(**encoded_inputs)
embeddings = outputs.last_hidden_state

# Calculate pairwise similarities between the MCQs
similarity_matrix = np.inner(embeddings.detach().numpy(), embeddings.detach().numpy())

# Reshape the similarity matrix dynamically
num_mcqs = len(mcqs)
similarity_matrix_reshaped = similarity_matrix.reshape(num_mcqs, -1)

# Perform k-means clustering on the reshaped similarity matrix
num_clusters = 3  # Adjust the number of clusters as desired
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
cluster_labels = kmeans.fit_predict(similarity_matrix_reshaped)

# Create a dictionary to map cluster labels to difficulty levels
difficulty_levels = {
    0: "Easy",
    1: "Medium",
    2: "Hard"
}
new_data=[]
# Print the cluster labels and their corresponding difficulty levels for each MCQ
for mcq, label in zip(mcqs, cluster_labels):
    difficulty_level = difficulty_levels[label]
    print(f"MCQ: {mcq}\nDifficulty Level: {difficulty_level}\n")


    new_data.append(difficulty_level)

df[column_name] = new_data

# Display the updated DataFrame
print(df[column_name])

# Calculate the silhouette score
silhouette_avg = silhouette_score(similarity_matrix_reshaped, cluster_labels)

print(f"Silhouette Score: {silhouette_avg}")    
# accuracy = accuracy_score(ground_truth_labels, predicted_difficulty_levels)
# print(f"Accuracy: {accuracy:.2f}")
# import pandas as pd  
# pd['Berries'] = df['4']
# pd.to_csv('/content/my2.csv', index=False)
# row0 = pd.next() 
# pd.insert(6, column = "Depwfewarrtment", value = df['7'])  
# pd.head()
# pd.to_excel('my4.csv')
```
